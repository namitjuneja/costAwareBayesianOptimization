This is a linear distribution. So no need to fit a straight line. 
Domains are:
ntasks [1,64] - what I need [2,32]
mem [2,128] - what I need [2,4,8,16,32]

```python
fig =plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(ntasks, mem, price)
ax.scatter(ntasks, mem, price)
ax.set_xlabel("ntasks")
ax.set_ylabel("mem")
ax.set_zlabel("price")
```
